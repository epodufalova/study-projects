# Exploratory Data Analysis


```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Country', data=df, hue='Country', palette='viridis', legend=False)
plt.title('Sales number by country')

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
```

*From the diagram above we can see which countries have the highest sales*

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Product', data=df, hue='Product', 
                   order=df['Product'].value_counts().index, 
                   palette='viridis', legend=False)
plt.title('Products popularity')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
```

*Here we can see which products are the most popular*

```python
plt.figure(figsize=(12, 6))
profit_by_country = df.groupby('Country')['Profit'].sum().sort_values(ascending=False)
ax = sns.barplot(x=profit_by_country.index, y=profit_by_country.values, 
                 hue=profit_by_country.index, palette='viridis', legend=False)
plt.title('Profit by country')
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='Discount Band', data=df, hue='Discount Band',
                   order=df['Discount Band'].value_counts().index,
                   palette='viridis', legend=False)
plt.title('Discounts analysis')
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
```
